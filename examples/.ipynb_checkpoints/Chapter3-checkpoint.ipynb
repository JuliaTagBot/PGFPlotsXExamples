# 3 Step-by-Step Tutorials

Reproduces some plots in Chapter 3 of the PGFManual at http://pgfplots.sourceforge.net/pgfplots_unstable.pdf

```julia
import PGFPlotsX
using LaTeXStrings
```

## 3.2 Solving a Real Use Case: Function Visualization

```julia
axis = PGFPlotsX.Axis(
    "title"  => "Inv. cum normal",
    "xlabel" => L"x",
    "ylabel" => L"y",
    "ymin" => -3,
    "ymax" => 3,
    "minor y tick num" => 1)

table = PGFPlotsX.PGFTable("plotdata/invcum.dat")

plot = PGFPlotsX.Plot(table, "blue"; incremental = false)
push!(axis, plot)
axis
```

```julia
axis = PGFPlotsX.Axis("axis lines" => "left",
                       "scaled ticks" => false,
                       "xticklabel style" => [
                           "rotate" => 90,
                           "anchor" => "east",
                           "/pgf/number format/precision" => 3,
                           "/pgf/number format/fixed",
                           "/pgf/number format/fixed zerofill"
                        ]
                    )

μ = 0
σ = 1e-3
expr = PGFPlotsX.PGFFunction("exp(-(x-$μ)^2 / (2 * $σ^2)) / ($σ * sqrt(2*pi))")

plot = PGFPlotsX.Plot(expr, 
    "red", 
    "domain" => "-3*$σ:3*$σ",
    "samples" => 201; 
    incremental = false)

push!(axis, plot)
axis
```

## 3.3 Solving a Real Use Case: Scientific Data Analysis

```julia
push!(PGFPlotsX.PGFPLOTS_CUSTOM_PREAMBLE, "\\usepackage{pgfplotstable}")
axis = PGFPlotsX.Axis("title" => "Convergence Plot",
                      "xlabel" => "Degrees of freedom",
                      "ylabel" => L"$L_2$ Error",
                      "xmode" => "log",
                      "ymode" => "log",
                      "grid" => "major")

for i in 2:4
    table = PGFPlotsX.PGFTable("plotdata/data_d$i.dat")
    plot = PGFPlotsX.Plot(table; label ="\$d = $i\$")
    push!(axis, plot)
end

table = PGFPlotsX.PGFTable("plotdata/data_d4.dat", 
    "x" => "dof",
    "y" => [
        "create col/linear regression" =>
            ["y" => "l2_err",
             "variance list" => "1000, 800, 600, 500, 400, 200, 100"
            ]
          ])

push!(axis, PGFPlotsX.Plot(table))

axis
```

## 3.4 Use Cases involving Scatter Plots

```julia
axis = PGFPlotsX.Axis("enlargelimits" => false, "colorbar")
table = PGFPlotsX.PGFTable("plotdata/concat_VV_together_grid.dat")
plot = PGFPlotsX.Plot(table, "only marks", 
                             "scatter",
                             "point meta" => "ln(1e-6+abs(\\thisrow{f(x)}))/ln(10)",
                             "mark size" => "0.6pt")
push!(axis, plot)
axis
```
