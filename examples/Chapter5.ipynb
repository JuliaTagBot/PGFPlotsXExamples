## 5.8 Grouping plots

```julia
cs = [[(0,0), (1,1), (2,2)],
      [(0,2), (1,1), (2,0)],
      [(0,2), (1,1), (2,1)],
      [(0,2), (1,1), (1,0)]]

@pgf gp = GroupPlot({group_style = { group_size = "2 by 2",}, height = "6cm", width = "6cm"})

for coords in cs
    push!(gp, Plot(Coordinates(coords)))
end

gp
```

```julia
@pgf begin
gp = GroupPlot(
    {
        group_style = 
        {
            group_size = "2 by 2",
            horizontal_sep = "0pt",
            vertical_sep = "0pt",
            xticklabels_at = "edge bottom"
        },
        xmin = 0,
        ymin = 0,
        height = "3.7cm",
        width = "4cm",
        no_markers
    })

p = Plot(Table("plotdata/group-1.dat"), { very_thick }; incremental = false)

push!(gp, "", "group/empty plot")
push!(gp, p, { xmin = 5, xmax = 10 , ymin = 50, ymax = 100 })
push!(gp, p, { xmax = 5, ymax = 50 })
push!(gp, p, { xmin = 5, xmax = 10 , ymax = 50, yticklabels = ""})

end

gp
```
